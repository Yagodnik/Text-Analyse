```python
try:
    file = open("text.txt", mode="r", encoding="utf-8")
    
    content = file.read()
except:
    print("Cant open file!")

file.close()
```

```python
content = content.replace(".", "")
content = content.replace(" ", "")
content = content.replace("!", "")
content = content.replace("?", "")
content = content.replace(",", "")
content = content.replace(":", "")
content = content.replace("â€”", "")
content = content.replace("(", "")
content = content.replace(")", "")
content = content.replace(";", "")
content = content.lower()
```

```python
class Symbol:
    def __init__(self):
        self.char = ""
        self.count = 0
```

```python
symbols = []

def contain_symbol(arr, char):
    for j in range(len(arr)):
        symbol = arr[j]

        if symbol.char == char:
            return j

    return -1

total_count = len(content)
for i in range(total_count):
    char = content[i]

    index = contain_symbol(symbols, char)
    if index > -1:
        symbols[index].count += 1
    else:
        symbol = Symbol()
        symbol.char = char
        symbol.count = 1

        symbols.append(symbol)
```

```python
symbols = sorted(symbols, key=lambda x: x.count, reverse=True)

print(total_count)
for i in range((len(symbols))):
    print(symbols[i].char + " / " + str(symbols[i].count))
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1)

data = [[] for i in range(len(symbols))]

for i in range(len(symbols)):
    freq = symbols[i].count / total_count
    data[i] = [symbols[i].char, symbols[i].count, f'{freq:0.3f}']

column_labels=["Symbols", "Count", "Freq"]
ax.axis('tight')
ax.axis('off')
ax.table(cellText=data, colLabels=column_labels, loc="center")

plt.show()
```
